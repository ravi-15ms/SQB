```r
 #For estimating lambda, let's first create a vector with random numbers
#generated by a exponential distribution with lambda = 34

x<-rexp(500,rate=10)#exponential siatribution to generate 5 random numbers with rate=20
rate<-10
# given that the generative distribution is 
# f(x) = lambda*exp(-x*lambda),

# the log-likelihood function will be 
# g(lambda,x) = length(x)*ln(lambda) - sum(x)*lambda

# by doing dg/dlambda = 0, we arrive in lambda = length(x)/sum(x) = 1/mean(x)
lambda = 1/mean(x)#actual MLE

# the critical point should be a maximum if dg²/d²lambda < 0
# which corresponds to -1/lambda^2 < 0 and it true for any real
# lambda, so we don't have to test it :)

# plot of the likelihood function:
likelihood = function(lambda, x)
  length(x)*log(lambda) - sum(x)*lambda

l = seq(1, 2*round(max(lambda, rate)), length.out = 100) #sequence for different lambda's

plot(l, likelihood(l, x), type = "l", xlab = "lambda", ylab = "likelihood")
points(lambda, likelihood(lambda, x), pch = 15, col = "red")
points(rate, likelihood(rate, x), pch = 15, col="blue")
legend("bottomright", legend = c("real lambda", "ML lambda"), fill = c("red", "blue"))

#for generating the vertical line
abline(v=lambda[which.max(likelihood(lambda,x))],col='red',lwd=3,lty=2) #MLE of lambda
abline(v=rate,col='blue',lwd=3,lty=2) #Set lambda

```
